```python
import numpy as np

mu, sigma = 0, 1 # mean and standard deviation
s = np.random.normal(mu, sigma, 100)

# grade igual à da figura
x = np.arange(0, 100)       # 0, 1, 2, …, 100
y = 10 * np.sin(x / 10) + s

```
