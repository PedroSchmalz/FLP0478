### Figures e Subplots

<p align="justify"> No Matplotlib, as "figures" e "subplots" são conceitos fundamentais para criar e organizar visualizações de dados de maneira eficaz. Eles permitem controlar como os gráficos são organizados em uma área de plotagem, o que é especialmente útil quando você deseja criar várias visualizações em uma única figura.




**Figure**

<p align="justify"> Uma "figure" é uma janela ou área de plotagem vazia onde você pode criar gráficos. É o recipiente principal para todos os elementos visuais, como gráficos, eixos, rótulos e legendas. Ao criar uma figura, você reserva uma área onde os gráficos serão desenhados. Você pode pensar em uma figura como uma tela em branco onde você pode adicionar visualizações.

Para criar uma figura no Matplotlib, você pode usar a função plt.figure(). Por exemplo:

````python
import matplotlib.pyplot as plt

# Criar uma figura
fig = plt.figure()
````

Até o momento não temos nada, só um objeto 'Figure' vazio. Para que possamos imprimir as 'telas em branco', precisamos dos subplots.

**Subplots**

<p align="justify"> Os "subplots" são as áreas dentro de uma figura onde os gráficos individuais são desenhados. Uma figura pode conter vários subplots organizados em uma grade. Cada subplot é identificado por um número de índice, começando do canto superior esquerdo e aumentando da esquerda para a direita e de cima para baixo.

<p align="justify"> Para criar subplots em uma figura, você pode usar a função plt.subplots(). Ela retorna tanto a figura quanto uma matriz de eixos (subplots) que você pode usar para criar gráficos em cada subplot. Aqui está um exemplo de como criar uma figura com 2 subplots:

````python
# Criar uma figura com 2 subplots em uma grade de 1x2
fig, axs = plt.subplots(1, 2)
````

Vamos ir por partes, primeiro vamos criar uma figura e adicionar um subplot na célula 1x1 (Pense em uma matriz, seria como o elemento na primeira linha e primeira coluna)

````python
# Criar uma figura com uma grade de 2x2 e adicionar 3 subplots

fig = plt.figure()


# Subplot da célula (1,1)

ax1 = fig.add_subplot(2,2,1)

````

Agora, vamos adicionar um elemento na segunda coluna, primeira linha

````python
# Adicionando o subplot
ax2 = fig.add_subplot(2,2,2)

# Imprimindo a figura
fig
````

Agora, o elemento na segunda linha, primeira coluna

````python
# Adicionando o subplot

ax3 = fig.add_subplot(2,2,3)

# Visualizando a figura

fig
````

Esses são os blocos de construção básicos do Matplotlib. Entendendo eles, se torna mais fácil montar suas próprias visualizações posteriormente. Por exemplo, vamos adicionar uma visualização no terceiro quadrante.

````python
np.random.seed(seed = 42) # Estabelecendo um seed para que os valores sejam sempre os mesmos


ax3.plot(np.random.standard_normal(100).cumsum(), # Pense em que tipo de distribuição estamos gerando para essa variável
         color = 'black', # Definindo a cor da linha a ser plotada
         linestyle = 'dashed' # Criando uma linha tracejada

         )

fig
````

(Tome cuidado ao rodar essa última célula de código várias vezes, você pode acabar ficando com várias linhas tracejadas geradas).

Os eixos (ou axis) tem métodos específicos que podem gerar visualizações pré-definidas. Por exemplo, se eu quiser um histograma, posso adicioná-lo ao eixo 1 utilizando o método '.hist':

````python
y = np.random.standard_normal(100)

ax1.hist(y,
         bins = 20, # Número de colunas do histograma
         color = 'black',
         alpha = 0.3) # Transparência do histograma (vai de 0 - transparente - até 1 - intenso)

fig
````

Esses subplots suportam qualquer tipo de visualização. Por exemplo, se eu desejar um scatterplot, primeiro preciso de duas variáveis:

````python
# Gerando 'a'

a = np.arange(30)

# Gerando 'b'

b = np.random.standard_normal(30)

ax2.scatter(a, # Eixo X
            b) # Eixo Y

fig
````

#### Ajeitando o espaço entre os subplots

Você pode ajustar manualmente os espaços entre os subplots, caso deseje. Vamos pegar um exemplo com quatro subplots de histogramas:

````python
# Criando a figura e os subplots explicitamente

fig, axes = plt.subplots(2, # Número de Linhas
                          2, # Número de colunas
                          sharex = True, # Estabelecer que os plots tem o mesmo eixo X
                          sharey = True, # Os plots tem o mesmo eixo Y
                          )

# Criando as observações

obs = np.random.standard_normal(500)


# Percorrendo os subplots e preenchendo com as figuras
for i in range(2):
  for j in range(2):
    axes[i,j].hist(obs,
                   bins = 50,
                   color = 'black',
                   alpha = 0.5)

# Ajustando o espaço branco entre os subplots

fig.subplots_adjust(wspace = 0, # (width)
                    hspace = 0) # espaço horizontal
````

### Explícito e Implícito na API do Matplotlib


<p align="justify"> No contexto da API do Matplotlib, "explícito" e "implícito" referem-se à maneira como você cria e manipula objetos relacionados à plotagem, como figuras e eixos. Esses conceitos estão relacionados à escolha entre o estilo de programação orientado a objetos (explícito) e o estilo de programação funcional (implícito). Em suma, é uma questão de <b> estilo</b>.

<b> Estilo Explícito </b>

1.   No estilo explícito, você cria explicitamente objetos que representam figuras, eixos e outros elementos da plotagem.
2.   Você invoca métodos nos objetos criados para adicionar elementos à sua plotagem.
3. Esse estilo é mais modular e oferece um maior nível de controle sobre cada elemento da plotagem.
4. É geralmente preferível ao criar visualizações complexas ou quando você precisa manipular várias figuras e subplots.


Esse é o estilo que estamos usando nesta aula, e é preferível também para evitar problemas de gráficos que sobrescrevem outros, ou problemas de posicionamento das figuras nos eixos. Aqui está um exemplo:

````python
# Criar uma figura explicitamente
fig = plt.figure()

# Criar eixos dentro da figura
ax = fig.add_subplot(1, 1, 1)

# Criar dados
y = np.random.standard_normal(100)
x = np.arange(100)

# Plotar dados nos eixos
ax.plot(x, y)

# Personalizar a plotagem
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico de Linha')

# Mostrar a figura
plt.show()
````

**Estílo Implícito**

1. No estilo implícito, você não cria explicitamente objetos de figura e eixos, mas chama diretamente funções do Matplotlib para criar visualizações.

2. Esse estilo é mais compacto e conveniente para tarefas simples de plotagem.

3. É uma escolha comum quando você deseja criar rapidamente gráficos simples ou quando está trabalhando interativamente em um ambiente como o Jupyter Notebook.

Você pode ler mais sobre as diferenças de estilos na documentação do [matplotlib](https://matplotlib.org/stable/users/explain/api_interfaces.html#api-interfaces). Essa é uma diferença que não é comentada nos textos sobre visualização em Python, mas que pode afetar o aprendizado, já que os autores adotam um dos estilos sem deixar claro, e isso gera diferenças importantes no jeito que os códigos são escritos. Aqui está um exemplo de gráfico no estilo implícito.

````python
# Criar um gráfico de linha diretamente sem criar figuras e eixos explicitamente
plt.plot(x, y)

# Personalizar a plotagem
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico de Linha')

# Mostrar o gráfico
plt.show()
````

<p align="justify">Como você pode ver, o resultado final é o mesmo. No entanto, este modo utiliza muito menos código. Só que pode ficar mais complicado de fazer figuras com diversos subplots nesse estilo.

### Marcações, rótulos e legendas

<p align="justify">No Matplotlib, os "ticks" (marcadores), "labels" (rótulos) e "legends" (legendas) são elementos essenciais para tornar seus gráficos mais informativos e compreensíveis. Eles são usados para indicar valores nos eixos, fornecer informações contextuais e identificar séries de dados em gráficos. Vamos entender cada um desses elementos:




**Ticks** (Marcadores):

<p align="justify">Os "ticks" são as marcas ao longo dos eixos de um gráfico que indicam valores específicos.
Eles ajudam os leitores a identificar rapidamente os valores correspondentes nos eixos.
Você pode personalizar a localização dos ticks, definir seus rótulos e formatar sua aparência.
Para personalizar os ticks, você pode usar as funções xticks() e yticks():

````python
# Definir ticks personalizados no eixo x e y
plt.xticks([0, 1, 2, 3, 4], # Valores de X
           ['A', 'B', 'C', 'D', 'E']) # Ticks que substituirão os valores

plt.yticks([0, 1, 2, 3, 4], # Valores de Y
           ['Baixo', 'Médio', 'Alto', 'Muito Alto', 'Máximo']) # Ticks que substituirão os valores

# Mostrar o gráfico
plt.show()
````

Labels (Rótulos):

<p align="justify">Os "labels" são usados para fornecer informações sobre os eixos, incluindo as unidades dos dados representados.
Eles são essenciais para tornar seus gráficos compreensíveis.
Você pode adicionar rótulos aos eixos x e y usando as funções xlabel() e ylabel():

````python
# Adicionar rótulos aos eixos x e y
plt.xlabel('Tempo (em meses)')
plt.ylabel('Receita (em milhões de dólares)')

# Mostrar o gráfico
plt.show()
````

Legends (Legendas):

<p align="justify">As "legends" são usadas em gráficos para identificar séries de dados ou elementos individuais.
Elas são especialmente úteis em gráficos que representam várias séries de dados.
Você pode adicionar legendas a um gráfico usando a função legend() e especificando rótulos para cada série de dados quando você faz o gráfico.

````python
# Dados de exemplo
x = np.arange(5)
y1 = [10, 8, 12, 6, 9]
y2 = [8, 7, 11, 5, 10]

# Plotar duas séries de dados com rótulos
plt.plot(x, y1, label='Série A')
plt.plot(x, y2, label='Série B')

# Adicionar uma legenda
plt.legend()

# Mostrar o gráfico
plt.show()
````

<p align="justify">Esses elementos são fundamentais para a criação de gráficos bem documentados e informativos no Matplotlib. Eles tornam seus gráficos mais compreensíveis, permitindo que os leitores identifiquem os valores nos eixos, entendam o contexto dos dados e distingam diferentes séries de dados quando necessário. Juntando os três conceitos, poderíamos ter algo assim:

````python
# Dados de exemplo
x = [1, 2, 3, 4, 5]
y1 = [10, 8, 12, 6, 9]
y2 = [8, 7, 11, 5, 10]

# Plotar duas séries de dados com rótulos
plt.plot(x, y1, label='Série A')
plt.plot(x, y2, label='Série B', linestyle = 'dotted')

# Adicionar uma legenda
plt.legend()

# Estabelecendo os labes dos eixos X e Y
plt.xlabel('Tempo (em meses)')
plt.ylabel('Receita (em milhões de dólares)')

plt.xticks(x, ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']) # Ticks que substituirão os valores

# Mostrar o gráfico
plt.show()
````

<p align="justify">Também é possível alterar a posição da legenda. A padrão é a 'best', que a posiciona de forma a não ficar por cima dos dados. Por exemplo, se eu quiser colocar no canto superior esquerdo, posso definir a opção 'upper left'. Além disso, para colocar um título no gráfico, posso utilizar o plt.title():

````python
# Dados de exemplo
x = [1, 2, 3, 4, 5]
y1 = [10, 8, 12, 6, 9]
y2 = [8, 7, 11, 5, 10]

# Plotar duas séries de dados com rótulos
plt.plot(x, y1, label='Série A')
plt.plot(x, y2, label='Série B', linestyle = 'dotted')

# Adicionar uma legenda
plt.legend(loc = 'upper left') # Colocando no canto superior esquerdo

# Estabelecendo os labes dos eixos X e Y
plt.xlabel('Tempo (em meses)')
plt.ylabel('Receita (em milhões de dólares)')

plt.xticks(x, ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']) # Ticks que substituirão os valores

# Colocando título

plt.title('Receita das Empresas A e B')

# Mostrar o gráfico
plt.show()
````

### Anotações e desenhos

<p align="justify">As anotações e desenhos em subplots são técnicas importantes para adicionar informações extras aos seus gráficos no Matplotlib. Eles permitem realçar pontos de interesse, fornecer explicações adicionais e tornar seus gráficos mais informativos.

**Anotações**:

<p align="justify">As anotações são usadas para adicionar texto ou setas a pontos específicos em seus subplots para explicar ou destacar informações importantes. Para adicionar anotações a um subplot, você pode usar a função annotate().

<p align="justify">Aqui está um exemplo de como adicionar uma anotação a um ponto específico em um gráfico de dispersão:

````python
# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [10, 8, 12, 6, 9]

# Criar um gráfico de dispersão
plt.scatter(x, y)

# Adicionar uma anotação ao ponto (3, 12)
plt.annotate('Pico', xy=(3, 12), xytext=(3.5, 11),
             arrowprops=dict(arrowstyle='->', color='red'))

# Mostrar o gráfico
plt.show()
````

<p align="justify">Suponha que estamos trabalhando com dados de desempenho de carros em relação ao consumo de combustível, e queremos destacar um carro específico no gráfico com uma anotação. Vamos usar um conjunto de dados de carros disponível no Seaborn para isso:

````python
import seaborn as sns

# Carregar o conjunto de dados de carros do sns
carros = sns.load_dataset("mpg")

# Criar um gráfico de dispersão
plt.figure(figsize=(10, 6)) # Definindo o tamanho da figura
plt.scatter(x='weight',
            y='mpg',
            data=carros,
            alpha=0.7)


# Coordenadas do carro que queremos destacar
carro_destacado = carros[carros['name'] == 'chevrolet impala']

# Anotar o carro destacado com uma seta
plt.annotate('Chevrolet Impala', xy=(carro_destacado['weight'].values[0], # Pegando o X do carro
                                     carro_destacado['mpg'].values[0]), # Pegando o Y do carro
             xytext=(50, 25),  # Posição do texto de anotação
             textcoords='offset points',
             arrowprops=dict(arrowstyle='->', color='black'))


# Adicionar rótulos de eixos
plt.xlabel('Peso do Carro')
plt.ylabel('Consumo de Combustível (mpg)')
plt.title('Desempenho de Carros em Relação ao Peso e Consumo de Combustível')

# Mostrar o gráfico
plt.show()
````

**Desenhos**:

<p align="justify">Os desenhos permitem que você adicione formas, linhas ou outras marcas visuais aos seus subplots. Você pode usar funções como plot(), fill_between(), fill_betweenx(), axhline(), axvline() e outras para criar desenhos personalizados.

<p align="justify">Aqui está um exemplo de como adicionar uma linha horizontal e uma forma preenchida a um subplot:

````python
# Dados de exemplo
x = [1, 2, 3, 4, 5]
y = [10, 8, 12, 6, 9]

# Criar um gráfico de linhas
plt.plot(x, y)

# Adicionar uma linha horizontal em y=10
plt.axhline(y=10, color='red', linestyle='--', label='Limite')


# Adicionar uma legenda
plt.legend()

# Mostrar o gráfico
plt.show()
````

<p align="justify">Você pode também desenhar formas geométricas utilizando algumas presentes no Matplotlib. Vamos trabalhar o exemplo de Mckinney (2022, p. 295), começando pelo retângulo:

````python
fig, ax = plt.subplots()

retan = plt.Rectangle((0.2, 0.75),  # (x, y) - As coordenadas do canto inferior esquerdo do retângulo
                      0.4,          # Largura do retângulo
                      0.15,         # Altura do retângulo
                      color='black', # Cor do retângulo (preto)
                      alpha=0.3)     # Transparência do retângulo (0.3 ou 30%)

ax.add_patch(retan)
````

Agora, o polígono (triângulo):

````python
polig = plt.Polygon([  # Lista de coordenadas dos vértices do polígono
                     [0.15, 0.15], # Vértice 1
                     [0.35, 0.4], # Vértice 2
                     [0.2, 0.6]], # Vértice 3
                     color = 'green',
                     alpha = 0.5)

ax.add_patch(polig)

fig
````

Por fim, o círculo:

````python
circ = plt.Circle((0.7,0.2), # (x, y) - As coordenadas do centro do círculo
                  0.15, # Raio do círculo
                  color = 'blue',
                  alpha = 0.3)

ax.add_patch(circ)

fig
````

### Salvando figuras

<p align="justify">Salvar figuras geradas com o Matplotlib é uma etapa importante ao criar visualizações para uso posterior em relatórios, apresentações ou para compartilhar com outras pessoas. O Matplotlib oferece várias opções para salvar figuras em diferentes formatos e tamanhos. Além disso, você pode personalizar várias configurações, como DPI (dots per inch ou pontos por polegada), tamanho da figura e formato do arquivo. Aqui estão os passos para salvar figuras no Matplotlib:

### Como escolher o gráfico certo

<p align="justify">Escolher a visualização correta para seus dados é fundamental. A escolha da visualização apropriada depende dos tipos dos dados que você possui. Aqui estão algumas diretrizes gerais para ajudar a escolher a visualização correta com base no tipo de dados:

1. **Dados Unidimensionais** (Univariados):

* Histograma: <p align="justify">Use um histograma para representar a distribuição de dados contínuos ou discretos em uma única variável. Isso ajuda a entender a frequência das diferentes faixas de valores.
* Gráfico de Barras: <p align="justify">Use um gráfico de barras para representar a contagem ou frequência de categorias ou valores discretos em uma única variável.
* Gráficos de densidade.


2. **Dados Bidimensionais** (Bivariados):

*  Gráfico de Dispersão: <p align="justify">Use um gráfico de dispersão para mostrar a relação entre duas variáveis contínuas. Isso ajuda a identificar padrões de correlação ou tendências.
* Gráfico de Barras Empilhadas: <p align="justify">Use um gráfico de barras empilhadas para mostrar como uma variável é dividida em categorias com base em outra variável categórica.
* Gráfico de Linha: <p align="justify">Use um gráfico de linha para representar a tendência ao longo do tempo ou ordem de classificação de duas variáveis contínuas.


3. Dados Multidimensionais (Multivariados):

* Gráfico de Dispersão em 3D: <p align="justify">Quando você tem três variáveis contínuas, pode usar um gráfico de dispersão em 3D para mostrar como elas estão relacionadas em um espaço tridimensional.
* Gráfico de Matriz de Dispersão: <p align="justify">Para explorar a relação entre várias variáveis, um gráfico de matriz de dispersão permite traçar gráficos de dispersão para todas as combinações possíveis de variáveis em um conjunto de dados.

4. Dados Temporais ou de Séries Temporais:

* Gráfico de Linha: <p align="justify">Use um gráfico de linha para mostrar tendências ao longo do tempo ou ordem de classificação.
* Gráfico de Área Empilhada: <p align="justify">Para comparar a contribuição de várias categorias ao longo do tempo, um gráfico de área empilhada pode ser eficaz.
* Histograma Temporal: <p align="justify">Para entender a distribuição dos dados ao longo do tempo, um histograma temporal pode ser útil.

5. Dados Geoespaciais:

* Mapas: <p align="justify">Quando seus dados estão associados a locais geográficos, os mapas podem ser uma escolha poderosa. Use mapas de calor, mapas de bolhas ou mapas <i>cloropleth</i> para representar diferentes aspectos dos dados geoespaciais.

6. Dados Categóricos ou Nominais:

* Gráfico de Barras: <p align="justify">Um gráfico de barras é uma escolha comum para representar a distribuição de dados categóricos ou nominais.
* Gráfico de Pizza: <p align="justify">Use um gráfico de pizza para mostrar a proporção de cada categoria em um conjunto de dados. [Não recomendado]

<p align="justify">Lembre-se de que essas são apenas diretrizes gerais, e a escolha da visualização específica pode depender da natureza única de seus dados e das perguntas que você deseja responder. Às vezes, é necessário experimentar diferentes tipos de gráficos para encontrar o que melhor representa seus dados de forma clara e informativa. Além disso, a clareza e a simplicidade são essenciais; evite escolher visualizações excessivamente complexas que possam confundir o público alvo.

Um site que pode lhe ajudar com a escolha do gráfico certo é o [data-to-viz](https://www.data-to-viz.com/#explore).

####  Scatterplots (Gráficos de Dispersão)

<p align="justify">Um scatterplot (gráfico de dispersão) é uma das visualizações mais simples e poderosas para representar a relação entre duas variáveis. Ele é amplamente utilizado na análise exploratória de dados para identificar padrões, tendências, clusters e até mesmo outliers nos dados. Aqui, iremos trabalhar com o exemplo de Massaron e Boschetti (2018).

````python
from sklearn.datasets import make_blobs

D = make_blobs(n_samples = 100, # N° de observações
               n_features = 2, #número de features
               centers = 3, # Grupos
               random_state = 7 # Seed
               )

grupos = D[1]

coordenadas = D[0]
````

<p align="justify">Temos três grupos, então temos que plotar três gráficos diferentes dentro de um mesmo painel.

````python
plt.plot(coordenadas[grupos==0,0], # Valores do Eixo X (0)
         coordenadas[grupos==0,1], # Valores do eixo Y (1)
         'ys', # Cor amarela e símbolo quadrado y(yellow) e s(square)
         label = 'grupo 1'
         )

plt.plot(coordenadas[grupos==1,0], # Valores do Eixo X (0)
         coordenadas[grupos==1,1], # Valores do eixo Y (1)
         'm*', # Cor magenta e símbolo de estrelas
         label = 'grupo 2'
         )

plt.plot(coordenadas[grupos==2,0], # Valores do Eixo X (0)
         coordenadas[grupos==2,1], # Valores do eixo Y (1)
         'rD', # Cor vermelha e símbolo de diamante
         label = 'grupo 3'
         )

plt.ylim(-2,10) # Limite do eixo Y
plt.yticks([10,6,2,-2]) # Ticks do eixo Y
plt.xticks([-15,-5,5,-15]) # Ticks do eixo X
plt.grid() # Adiciona o grid

plt.annotate('Quadrados',
            (-12,2.5))

plt.annotate('Estrelas',
             (0,6))

plt.annotate('Diamantes', (8,3))

plt.legend(loc = 'lower left', numpoints=1)

plt.show()
````

#### Histogramas

<p align="justify">Um histograma é uma representação gráfica da distribuição de frequência de um conjunto de dados. É uma ferramenta valiosa para visualizar a distribuição de valores em um conjunto de dados, identificar tendências, padrões e características da distribuição dos dados. No Matplotlib, você pode criar histogramas usando a função plt.hist() ou o método ax.hist() de um objeto de eixo.

<p align="justify">Um histograma permite visualizar como os dados estão distribuídos em relação a uma variável específica. Isso pode ajudar a identificar se os dados são normalmente distribuídos, assimétricos (à esquerda ou à direita), multimodais (possuem múltiplos picos), etc.

````python
x = np.random.normal(loc = 0.0, scale = 1.0, size = 500)
z = np.random.normal(loc = 3.0, scale = 1.0, size = 500)

plt.hist(np.column_stack((x,z)),
         bins = 20,
         color = ['c', 'b'],
         stacked = True,
         alpha = 0.5)

plt.grid()
plt.show()
````

#### Gráficos de barras

<p align="justify">Gráficos de barras são uma das visualizações mais comuns e eficazes para representar dados categóricos. Eles permitem mostrar a contagem ou a frequência de ocorrência de diferentes categorias e comparar essas categorias de forma visual. No Matplotlib, você pode criar gráficos de barras usando a função plt.bar() ou o método ax.bar() de um objeto de eixo.

````python
# Dados fictícios para o exemplo
categorias = ["Categoria A", "Categoria B", "Categoria C", "Categoria D"]
medidas = [25, 40, 30, 55]

# Criar um gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(categorias, medidas, color='blue')

# Adicionar rótulos e título
plt.xlabel('Categorias')
plt.ylabel('Medidas')
plt.title('Gráfico de Barras Exemplo')

# Exibir o gráfico
plt.show()
````

<p align="justify">O exemplo de Massaron e Boschetti (2018, p. 291) ilustra também a possibilidade de utilizar gráficos de barras para demonstrar incerteza, além da diferença de utilizar gráficos barras horizontais e verticais.



````python
from sklearn.datasets import load_iris

# Carregando o banco e calculando os valores
iris = load_iris()
media = np.mean(iris.data, axis = 0)
desvio = np.std(iris.data, axis = 0)
range_ = range(np.shape(iris.data)[1])

# Fazendo o gráfico da primeira célula (Barras horizontais)

plt.subplot(1,2,1)
plt.title('Barras Horizontais')
plt.barh(range_, media, color = 'r',
         xerr = desvio, alpha = 0.4, align = 'center')
plt.yticks(range_, iris.feature_names)


# Gráfico da segunda célula (Barras Verticais)

plt.subplot(1,2,2)
plt.title("Barras verticais")
plt.bar(range_, media, color = 'b',
        yerr = desvio, alpha = 0.4, align = 'center')
plt.xticks(range_, range_)

# Gráfico final
plt.show()
````

<p align="justify">Há vários métodos embutidos no Pandas que permitem fazer visualizações de diversos tipos diretamente por ele. Por exemplo, voltando ao banco fictício criado anteriormente.

````python
# Criar um DataFrame fictício
num_funcionarios = 100

data = {
    'ID': range(1, num_funcionarios + 1),
    'Nome': ['Funcionário' + str(i) for i in range(1, num_funcionarios + 1)],
    'Idade': np.random.randint(20, 65, size=num_funcionarios),
    'Cargo': np.random.choice(['Analista', 'Gerente', 'Desenvolvedor', 'Designer'], size=num_funcionarios),
    'Salário (R$)': np.random.uniform(2500, 10000, size=num_funcionarios)
}

df = pd.DataFrame(data)


# Gráfico de barras para a distribuição de cargos dos funcionários

# Esta linha cria um gráfico de barras.
# 'df['Cargo']' seleciona a coluna 'Cargo' do DataFrame 'df', que contém os cargos dos funcionários.
# 'value_counts()' conta quantas vezes cada valor único aparece na coluna 'Cargo'.
# 'plot(kind='bar', color='lightcoral')' cria um gráfico de barras com a cor das barras definida como 'lightcoral'.
df['Cargo'].value_counts().plot(kind='bar', color='lightcoral')

# Adiciona um rótulo ao eixo x do gráfico
plt.xlabel('Cargo')

# Adiciona um rótulo ao eixo y do gráfico
plt.ylabel('Número de Funcionários')

# Adiciona um título ao gráfico
plt.title('Distribuição de Cargos dos Funcionários')

# Rotaciona os rótulos do eixo x em 45 graus para melhor legibilidade
plt.xticks(rotation=45)

# Exibe o gráfico
plt.show()
````

<p align="justify">Observe que o Pandas gera um objeto comum ao Matplotlib, que nos permite alterar o gráfico da mesma forma, adicionando nossas preferências de títulos, labels, etc. Se desejar um gráfico de barras horizontal, basta alterar a opção "kind" do plot.

````python
df['Cargo'].value_counts().plot(kind='barh', color='lightblue')

# Adiciona um rótulo ao eixo x do gráfico
plt.xlabel('N° de Funcionários')

# Adiciona um rótulo ao eixo y do gráfico
plt.ylabel('Cargo')

# Adiciona um título ao gráfico
plt.title('Distribuição de Cargos dos Funcionários')


# Exibe o gráfico
plt.show()
````

O conjunto de dados "Iris" é um dos conjuntos de dados mais famosos e amplamente utilizados na área de aprendizado de máquina e classificação. Ele foi introduzido pela primeira vez pelo estatístico britânico e biólogo Ronald A. Fisher em 1936 como parte de seu trabalho sobre discriminação linear. O conjunto de dados Iris é frequentemente utilizado para fins educacionais, benchmarking de algoritmos de classificação e demonstrações de técnicas de análise de dados.

<p align="justify">O conjunto de dados Iris consiste em 150 amostras de flores de íris, divididas em três espécies diferentes: setosa, versicolor e virginica. Cada espécie tem 50 amostras.

````python
from sklearn.datasets import load_iris

# carregando o banco
iris = load_iris()

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
groups = list(iris.target)

iris_df['groups'] = pd.Series([iris.target_names[k] for k in groups])

iris_df
````

<p align="justify">Novamente, vamos fazer um scatterplot, utilizando as funções do pandas para Data Frames:

````python
# Mapeando as cores

colors_palette = {0: 'red', 1: 'yellow', 2:'blue'}
colors = [colors_palette[c] for c in groups]

iris_df.plot(kind = 'scatter', x = 0, y = 1, c = colors, alpha = 0.7)
````

<p align="justify">Podemos transformar o scatterplot em um hexbin, uma forma alternativa de visualizar a mesma informação. No entanto, perdemos a informação por grupos.

````python
iris_df.plot(kind = 'hexbin', x = 0, y = 1, gridsize = 10)
````

<p align="justify">Uma análise exploratória interessante pode ser a de fazer uma matriz de scatterplots com gráficos de densidade das variáveis, permitindo vasculhar o banco inteiro e identificar possíveis correlações, assimetrias, etc.

````python
from pandas.plotting import scatter_matrix

# Mapeamento de cores
colors_palette = {0: 'red', 1: 'green', 2: 'blue'}
colors = [colors_palette[c] for c in groups]

# Fazendo a matriz de scatterplots

scatter_mat = scatter_matrix(iris_df, # Se você não salvar a matriz em um objeto, aparece um monte de informação de texto
               alpha = 0.4,
               figsize = (8,8),
               color = colors,
               diagonal = 'kde') # definindo que a diagonal da matriz vai ser uma kernel density
````

## Seaborn

<p align="justify">O Seaborn é uma biblioteca de visualização de dados em Python baseada no Matplotlib. O Seaborn é especialmente útil para a análise exploratória de dados e a visualização de padrões em conjuntos de dados complexos. Aqui estão algumas características e funcionalidades-chave do Seaborn:


* Estilo Atraente: <p align="justify">O Seaborn melhora a estética padrão dos gráficos em comparação com o Matplotlib, tornando-os mais atraentes e prontos para uso em apresentações e relatórios.

* Integração com Pandas: <p align="justify">O Seaborn é altamente compatível com Data Frames do Pandas. Isso facilita a criação de gráficos diretamente a partir dos dados em um Data Frame.

* Suporte para Plots Estatísticos: <p align="justify">O Seaborn inclui uma variedade de funções para criar gráficos estatísticos com facilidade, como gráficos de distribuição, boxplots, violinos, gráficos de barras e de dispersão com regressão linear incorporada.

* Paletas de Cores: <p align="justify">O Seaborn oferece uma ampla gama de paletas de cores pré-definidas para melhorar a distinção de cores em seus gráficos. Você pode personalizar as paletas de cores ou criar as suas próprias.

* Facetamento: <p align="justify">O Seaborn permite criar subplots (facetamento) com base em categorias em seus dados, o que é útil para visualizar relações em subgrupos de dados.

* Manipulação de Escala: <p align="justify">O Seaborn facilita a manipulação de escalas em gráficos, como escala logarítmica, para melhor representação dos dados.

* Estimadores Estatísticos: <p align="justify">É possível adicionar estimativas estatísticas, como média ou mediana, diretamente aos gráficos de dispersão para uma análise mais detalhada.

* Personalização: <p align="justify">Embora o Seaborn ofereça estilos e paletas de cores prontas para uso, você pode personalizar quase todos os aspectos de seus gráficos, incluindo títulos, rótulos, cores e mais.

<p align="justify">Aqui está um exemplo simples de código que utiliza o Seaborn para criar um gráfico de dispersão com uma linha de regressão:

````python
import seaborn as sns

# Carregue um DataFrame fictício
import pandas as pd
data = {'X': [1, 2, 3, 4, 5], 'Y': [2, 4, 5, 4, 6]}
df = pd.DataFrame(data)

# Crie um gráfico de dispersão com regressão linear
sns.regplot(x='X', y='Y', data=df)
plt.title('Gráfico de Dispersão com Regressão Linear')
plt.show()
````

### Split, Apply, Combine (SAC)

<p align="justify"> O conceito de "split, apply e combine" é fundamental em computação e análise de dados, frequentemente utilizado em linguagens de programação como Python para processar e analisar conjuntos de dados. Esse padrão é frequentemente associado à biblioteca pandas em Python, mas sua aplicação não se limita a essa linguagem.

**Split** (Dividir):

<p align="justify"> A fase de "split" envolve a divisão de um conjunto de dados maior em grupos menores ou subconjuntos com base em algum critério específico. Esse critério pode ser uma coluna em um dataframe, um atributo de classificação, uma condição lógica, ou qualquer outra coisa que permita a divisão dos dados em partes distintas.

**Apply** (Aplicar):

A fase "apply" envolve a aplicação de uma operação ou função a cada um dos grupos ou subconjuntos de dados criados na etapa de "split". Essa operação pode ser uma agregação estatística, uma transformação de dados, uma análise personalizada ou qualquer outra manipulação que seja relevante para o problema em questão.


**Combine** (Combinar):

Após a aplicação da operação em cada grupo durante a etapa "apply", os resultados individuais são combinados novamente em um único resultado global. Isso pode envolver a combinação de estatísticas agregadas, a concatenação de dados transformados ou qualquer outra operação que agregue os resultados dos grupos em um resultado geral.

Em resumo, o padrão "split, apply e combine" é uma abordagem sistemática para dividir um problema de processamento ou análise de dados em etapas gerenciáveis. Ele permite que você lide com conjuntos de dados grandes e complexos, dividindo-os em partes menores, aplicando operações relevantes a essas partes e, finalmente, combinando os resultados em uma visão geral ou conclusão. ### Método '.groupby'


 O método .groupby() no pandas é usado para agrupar um DataFrame com base nos valores de uma ou mais colunas. Esse método permite que você divida seus dados em grupos com base em critérios específicos e, em seguida, aplique funções de agregação ou análise a esses grupos.

 Aqui está um exemplo de como o método .groupby() funciona:

````python
# Criar um DataFrame de exemplo
data = {'Grupo': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Valor': [10, 20, 15, 25, 12, 30]}

df = pd.DataFrame(data)

# Agrupar o DataFrame por 'Grupo'
grupo_df = df.groupby('Grupo')

# Calcular a média dos valores em cada grupo
media_por_grupo = grupo_df['Valor'].mean()

print(media_por_grupo)
````

Agora vamos entender como funciona esse método. Primeiro, vamos analisar o resultado do agrupamento

````python
# Agrupar o DataFrame por 'Grupo'
grupo_df = df.groupby('Grupo')

grupo_df
````

Perceba que só temos um objeto de agrupamento do Pandas, esse é o primeiro passo do Split, Apply e Combine: dividimos o banco de dados de acordo com a coluna de "Grupo". Depois, aplicamos o método de cálculo da média dos valores por grupo com '.mean()'. O método .groupby() é uma ferramenta poderosa no pandas para análise de dados, pois permite que você realize operações em grupos de dados com base em critérios específicos, facilitando a compreensão e a manipulação de conjuntos de dados complexos. ### Agregação dos dados

O método .groupby junta os dados de acordo com algum grupo/categoria. No entanto, ele não faz nada com eles ainda. Na seção anterior, acabamos adiantando um pouco da agregação dos dados, ou resumo, com o método .mean() após o agrupamento. A agregação dos dados envolve qualquer método que resume as informações individuais das observações em uma informação 'escalar' (ou única), como a média (mesmo que seja as médias dos grupos). Existem vários métodos de agregação de dados que já são otimizados e adequados para os Pandas, mas nada nos impede de criarmos funções próprias que podemos aplicar nos dados se assim o desejarmos. a Tabela 10-1 do texto de Mckinney (2022, p. 329) mostra uma lista dos principais métodos existentes. No entanto, vamos trabalhar alguns deles aqui.

Voltando para nosso exemplo anterior, podemos querer descobrir quais são os menos valores por grupo:

````python
# Criar um DataFrame de exemplo
data = {'Grupo': ['A', 'B', 'A', 'B', 'A', 'B'],
        'Valor': [10, 20, 15, 25, 12, 30]}

df = pd.DataFrame(data)

# Agrupando os dados
grupo = df.groupby('Grupo')


# Aplicando a função nsmallest
grupo['Valor'].nsmallest(1) # Somente o menor número


````

Para aplicar suas próprias funções, pode utilizar o método '.aggregate', que aplica a função de acordo com os grupos do '.groupby()'. Por exemplo, com base em Mckinney, poderíamos querer saber a diferença entre os maiores e menores valores por grupo


````python
# Função personalizada para calcular a diferença entre o maior e o menor valor
def menos_e_mais(arr):
    return arr.max() - arr.min()

resultado = grupo.agg(menos_e_mais)


resultado
````

````python
resultado = resultado.reset_index()

sns.barplot(y = 'Valor', x = 'Grupo', data=resultado, palette='viridis')
plt.title('Diferença entre o maior e menor valor')
plt.show()
````

### Apply

O método apply() em Python é usado principalmente em Pandas para aplicar uma função personalizada a uma série (ou coluna) ou DataFrame, ao longo de um eixo específico. Ele é uma ferramenta poderosa para transformação de dados e é amplamente utilizado na limpeza, preparação e análise de dados.

Aqui está a sintaxe básica do método apply():



```
DataFrame.apply(func, axis=0)
```



func: A função personalizada que você deseja aplicar aos dados.
axis: O eixo ao longo do qual a função será aplicada (0 para colunas ou 1 para linhas).

A função personalizada passada para apply() deve ser capaz de lidar com um único valor ou uma série de valores, dependendo de como você a está aplicando. Vou fornecer exemplos para ilustrar o uso do apply() em diferentes cenários:

````python
# Criar um DataFrame de exemplo
data = {'Valores': [1, 2, 3, 4, 5]}

df = pd.DataFrame(data)

# Definir uma função personalizada para elevar ao quadrado
def square(x):
    return x ** 2

# Aplicar a função à coluna 'Valores' usando apply
df['Valores ao Quadrado'] = df['Valores'].apply(square)

print(df)
````

Os usos mais básicos do apply acabam sendo cobertos pelos métodos padrões do Pandas, como mean(), describe(), etc. No entanto, com um pouco de criatividade e dependendo da sua necessidade, pode ser um método extremamente útil. Não se esqueça de sempre consultar a [documentação](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) de cada biblioteca sobre os métodos e funções, pois podem ser extremamente úteis e abrir alternativas que não havia pensado antes.

### Pivot-tables e Cross-Tabulation

As funções pivot_table e crosstab são ferramentas poderosas no Pandas para realizar análises e resumos de dados em formatos tabulares. Elas permitem que você organize e resuma seus dados de diferentes maneiras para obter insights e informações úteis.


**Pivot-Table**

A função pivot_table permite que você crie tabelas dinâmicas semelhantes às que são encontradas em planilhas, onde você pode resumir e agregar dados com base em critérios específicos. Ela é útil quando você deseja transformar um DataFrame em um formato que seja mais adequado para análise. A sintaxe básica da pivot_table é a seguinte:



```
pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean')
```



* data: O DataFrame de origem.
* values: A coluna ou colunas que você deseja resumir.
* index: As colunas que serão usadas como índice da tabela pivô.
* columns: As colunas que serão usadas como cabeçalhos de coluna na tabela pivô.
* aggfunc: A função de agregação a ser aplicada aos valores quando ocorrerem várias correspondências para uma combinação de índices e colunas.

Exemplo de uso:

````python
import pandas as pd

data = {'Produto': ['A', 'B', 'A', 'B', 'A'],
        'Vendas': [100, 200, 150, 250, 120],
        'Região': ['Norte', 'Sul', 'Norte', 'Sul', 'Norte']}

df = pd.DataFrame(data)

# Criar uma tabela pivô para resumir as vendas por Produto e Região
pivot = pd.pivot_table(df, values='Vendas', index='Produto', columns='Região', aggfunc='sum')

print(pivot)
````

**Cross-tabulation**

A função crosstab (tabela de contingência) é usada para calcular uma tabela de frequência de duas ou mais variáveis categóricas. Ela é útil para entender a relação entre diferentes categorias de variáveis e pode ser usada para criar tabelas cruzadas. A sintaxe básica da crosstab é a seguinte:



```
pd.crosstab(index, columns, values=None, aggfunc=None, rownames=None, colnames=None)

```



* index: A variável a ser usada no eixo vertical (linhas).
* columns: A variável a ser usada no eixo horizontal (colunas).
* values: A coluna que contém os valores a serem agregados (opcional).
* aggfunc: A função de agregação a ser aplicada aos valores (opcional).
rownames e colnames: Nomes opcionais para as linhas e colunas na tabela cruzada.

Exemplo de uso da crosstab:

````python
import pandas as pd

data = {'Gênero': ['Masculino', 'Feminino', 'Masculino', 'Feminino', 'Masculino'],
        'Estado Civil': ['Casado', 'Solteiro', 'Solteiro', 'Casado', 'Casado']}

df = pd.DataFrame(data)

# Criar uma tabela cruzada para contar o número de pessoas em cada categoria de Gênero e Estado Civil
cross_tab = pd.crosstab(df['Gênero'], df['Estado Civil'])

print(cross_tab)
````

Este exemplo cria uma tabela cruzada que mostra o número de pessoas em cada categoria de Gênero e Estado Civil.
